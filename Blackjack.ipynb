```python
from IPython.display import clear_output
import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','Ace')
values = {'Two' : 2, 'Three' : 3, 'Four' : 4, 'Five' : 5, 'Six' : 6, 'Seven' : 7,'Eight' : 8, 'Nine' : 9, 'Ten' : 10,
          'Jack':10 , 'Queen' : 10, 'King' : 10, 'Ace' : 11
         }

game_on = True
```

```python
class Cards:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    # To print what is in an object of the Cards class
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = [] # To start with an empty deck
        for suit in suits:
            for rank in ranks:
                self.deck.append(Cards(suit,rank))
                
   # To print what is in an object of the Deck class
    def __str__(self):
        cards_in_deck = " "
        for card in self.deck:
            cards_in_deck += "\n "+card.__str__()
        return cards_in_deck
    
   # To shuffle the cards
    def shuffle_cards(self):
        random.shuffle(self.deck)
        
   # To deal  
    def deal(self):
        card = self.deck.pop()
        return card
    
    
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def __repr__(self):
        return str(self.cards)
        
        
    # To add cards to the player's and dealer's hands    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces +=1
    
    # To assign a value for ace in the best interest of the player     
    def assign_value_ace(self):
        if self.value > 21 and self.aces :
            self.value -= 10
            self.aces -=1
```

```python
class Bet:
    
    def __init__(self,bet,total):
        self.bet = bet
        self.total = total
    
    # To calculate and display the chips that the player has won    
    def winning_bet(self):
        self.total += self.bet*2
        print("The player has won ",self.total," chips.")
    
    # To calculate and display the chips that the player has lost    
    def losing_bet(self):
        self.total = 0 - self.bet
        print("The player has lost ",self.total," chips.")
        
```

```python
# To let the player and dealer take cards
def take_a_card(deck,hand) :
    
    hand.add_card(deck.deal())
    hand.assign_value_ace()
```

```python
# To display the cards of the player and the dealer while the game is on. Only one card of the dealer is displayed , and the other is hidden.
def show_cards_gameon(dealer,player):
    clear_output()
    
    print("The Dealer's cards are as follows ")
    print("First card : Is Hidden ")
    print("Second Card ",dealer.cards[1])
    print("The Players cards are as follows ")
    for card in player.cards:
        print(card)
    
```

```python
# To display the cards of the player and the dealer while the game is not on. All the cards of the player and the dealer are displayed now
def show_cards_gameoff(dealer,player):
    
    print("The Dealer's cards are as follows ")
    for card in dealer.cards:
      print(card)
    print("The Players cards are as follows ")
    for card in player.cards:
      print(card)
```

```python
# To let the player choose to hit or to stand
def hit_or_stand (deck,hand):
    global playing
    while True:
        choice = input("Enter h to hit or s to stand ")
        if choice.lower() == 'h':
            take_a_card(deck,hand)
        elif choice.lower() == 's':
            playing = False
        else:
            print("Please enter a valid input ")
            continue
        break
```

```python
def player_busts(player_bet):
    clear_output()
    print("*************************")
    print ("The player has busted! ")
    print("*************************")
    player_bet.losing_bet()
    
def player_wins(player_bet):
    clear_output()
    print("*************************")
    print("The player has won! ")
    print("*************************")
    player_bet.winning_bet()
    
def dealer_busts(player_bet):
    clear_output()
    print("*************************")
    print("The dealer has busted!")
    print("*************************")
    player_bet.winning_bet()
    
def dealer_wins(player_bet):
    clear_output()
    print("*************************")
    print("The dealer has won ")
    print("*************************")
    player_bet.losing_bet()
    
def push():
    clear_output()
    print("*************************")
    print("Its a push ")
    print("*************************")
```

```python
#game_on = True
#while game_on == True:

print("Welcome to the Blackjack Game !!!!! ")
deck = Deck() # To create an object of the Deck class
deck.shuffle_cards() # To shuffle the deck

hand_of_player = Hand() # To create an object of the Hand class for the player
hand_of_player.add_card((deck.deal())) # To add the first card to the hand of the player
hand_of_player.add_card((deck.deal())) # To add the second card to the hand of the player

hand_of_dealer = Hand() # To create an object of the Hand class for the dealer
hand_of_dealer.add_card((deck.deal())) # To add the first card to the hand of the dealer
hand_of_dealer.add_card((deck.deal())) # To add the second card to the hand of the dealer


# To take a bet from the player
bet_amount= int(input("How many chips do you want to bet?"))
player_bet = Bet(bet_amount,0) # To create an object of the Bet class for the Player


show_cards_gameon(hand_of_dealer,hand_of_player) # To display the hand of the player and the hand of the dealer
playing = True

while playing:

    hit_or_stand(deck,hand_of_player) # To check if the player wants to hit or stand

    show_cards_gameon(hand_of_dealer,hand_of_player) # To display the hand of the player and the hand of the dealer

    if hand_of_player.value == 21 : # To check if the player has won
        player_wins(player_bet)
        show_cards_gameoff(hand_of_dealer,hand_of_player)

    elif hand_of_player.value > 21 : # To check if the player has busted
        player_busts(player_bet)
        show_cards_gameoff(hand_of_dealer,hand_of_player)
        break

if hand_of_player.value <= 21:

    while hand_of_dealer.value < 17: # To get the dealer to take cards till the value of his hand reaches atleast 17
        take_a_card(deck,hand_of_dealer)


    if hand_of_dealer.value > 21: # To check if the dealer has busted
        dealer_busts(player_bet)
        show_cards_gameoff(hand_of_dealer,hand_of_player)


    elif hand_of_dealer.value > hand_of_player.value : # To check if the dealer has won
        dealer_wins(player_bet)
        show_cards_gameoff(hand_of_dealer,hand_of_player)

    elif hand_of_player.value > hand_of_dealer.value : # To check if the player has won
        player_wins(player_bet)
        show_cards_gameoff(hand_of_dealer,hand_of_player)

    else: # When neither players have won and it is a push
        push()
        show_cards_gameoff(hand_of_dealer,hand_of_player)
```
